# Company Sales Graph (Matplotlib Excercise)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data/company_sales_data.csv')
df
```

#### Read Total profit of all months and show it using a line plot

```python
plt.title("Company Profit per Month")

plt.xlabel("Month Number")
month_number = df['month_number']

plt.ylabel("Total Profit")
total_profit = df['total_profit']

plt.plot(month_number, total_profit)

plt.show()
```

#### Get total profit of all months and show line plot with the following Style properties

```python
plt.title("Company Sales data of last year")

plt.xlabel("Month Number")
month_number = df['month_number']

plt.ylabel("Solds units number")
sales_units = df['total_units']

plt.plot(month_number, sales_units, '--r', linewidth=3)
plt.plot(month_number, sales_units, 'o', color='black', markersize=8)

plt.legend(['Profit data of last year'], loc = 'lower right')

plt.show()
```

#### Read all product sales data and show it using a multiline plot

```python
plt.title("Sales Data")

# X axis
plt.xlabel("Month Number")
month_number = df['month_number']

# Y axis
plt.ylabel("Sales units number")
sales_units = df['total_units']

# data
face_cream_sales = df['facecream'].tolist()
face_wash_sales = df['facewash'].tolist()
toothpaste_sales = df['toothpaste'].tolist()
bath_soap_sales = df['bathingsoap'].tolist()
shampoo_sales = df['shampoo'].tolist()
moisturizer_sales = df['moisturizer'].tolist()

# Plot the data
plt.plot(month_number, face_cream_sales, marker='o', linewidth=3, label = 'Face Cream Sales Data')
plt.plot(month_number, face_wash_sales, marker='o', linewidth=3, label = 'Face Wash Sales Data')
plt.plot(month_number, toothpaste_sales, marker='o', linewidth=3, label = 'Toothpaste Sales Data')
plt.plot(month_number, bath_soap_sales, marker='o', linewidth=3, label = 'Bath Soap Sales Data')
plt.plot(month_number, shampoo_sales, marker='o', linewidth=3, label = 'Shampoo Sales Data')
plt.plot(month_number, moisturizer_sales, marker='o', linewidth=3, label = 'Moisturizer Sales Data')


# Legend
plt.legend(loc = 'upper left')
```

#### Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.title("Tooth Paste Sales Data")

# X Axis
plt.xlabel("Month Number")
month_number = df['month_number'].tolist()

# Y Axis
plt.ylabel("Number of Units Sold")
tooth_paste_sales = df['toothpaste'].tolist()

# Plot the data
plt.scatter(month_number, tooth_paste_sales, marker='o', color='blue', s=100, label = 'Tooth Paste Sales Data')
plt.grid(True, linewidth=1, linestyle='--')
plt.legend(loc = 'upper left')
```

#### Read Face Cream and Face Wash product sales data and show it using the bar chart

```python
plt.title("Facewash and Facecream Sales Data")
plt.xlabel("Month Number")
month_number = df['month_number'].tolist()
plt.ylabel("Sales Units in Numbers")

# Data
facewash_sales = df['facewash'].tolist()
facecream_sales = df['facecream'].tolist()

plt.bar([x - 0.25 for x in month_number], face_wash_sales, width = 0.25, label = 'Face Wash Sales Data', color = 'red')
plt.bar([x + 0.25 for x in month_number], face_cream_sales, width = 0.25, label = 'Face Cream Sales Data', color = 'blue')

plt.legend(loc = 'lower right')
plt.grid(True, linewidth=1, linestyle='--')
plt.show()
```

#### Read sales data of bathing soap of all months and show it using the pie chart

```python
plt.title("Bathing Soap Sales Data")
plt.xlabel("Month Number")
month_number = df['month_number'].tolist()
plt.ylabel("Sales Units in Numbers")

# data
bathingsoap_sales = df['bathingsoap'].tolist()

plt.bar([x - 0.5 for x in month_number], bathingsoap_sales, width = 0.5, label = 'Bathing Soap Sales Data', color = 'red')
plt.grid(True, linewidth=1, linestyle='--')
plt.legend(loc = 'upper left')
plt.show()
```

#### Calculate total sale data for last year for each product and show it using a Pie chart

```python
plt.title("Sales Data")

month_number = df['month_number'].tolist()
labels = ['Face Wash', 'Face Cream', 'Tooth Paste', 'Bathing Soap', 'Shampoo', 'Moisturizer']
salesData = [
    df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(), df['bathingsoap'].sum(), df['shampoo'].sum(), df['moisturizer'].sum()
]

plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%1.1f%%', shadow=True)
plt.legend(loc='lower right')

plt.show()
```

#### Read Bathing soap facewawh of all months and display it using the Subplot

```python
bath_soap_sales = df['bathingsoap'].tolist()
face_wash_sales = df['facewash'].tolist()

f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(month_number, bath_soap_sales, color="k", label='Bathing Soap Sales Data', marker='o', linewidth=3)
axarr[0].set_title('Bathing Soap Sales Data')

axarr[1].plot(month_number, face_wash_sales, color="b", label='Face Wash Sales Data', marker='o', linewidth=3)
axarr[1].set_title('Face Wash Sales Data')

plt.xlabel("Month Number")
plt.ylabel("Sales Units in Numbers")

plt.show()
```

#### Read all product sales data and show it using the stack plot

```python
plt.title("All Product Sales Data")
plt.xlabel("Month Number")
month_number = df['month_number'].tolist()
plt.ylabel("Sales Units in Numbers")

# data
face_cream_sales = df['facecream'].tolist()
face_wash_sales = df['facewash'].tolist()
tooth_paste_sales = df['toothpaste'].tolist()
bath_soap_sales = df['bathingsoap'].tolist()
shampoo_sales = df['shampoo'].tolist()
moisturizer_sales = df['moisturizer'].tolist()

# plotting data
plt.stackplot([x - 0.25 for x in month_number], face_wash_sales, face_cream_sales, tooth_paste_sales, bath_soap_sales, shampoo_sales, moisturizer_sales, colors=['m', 'c', 'r', 'k', 'y', 'g'], labels=['Face Wash', 'Face Cream', 'Tooth Paste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])

plt.legend(loc='upper left')
```
